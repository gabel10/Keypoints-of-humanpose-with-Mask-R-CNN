## Train / Val data augmentation

Para realizar la división de los datos, primero tenemos que tener lo siguiente en formato de carpetas:
* Images. carpeta donde están las imágenes en formato JPG de tamaño (640x640)
* Masks. carpeta donde están las máscaras en formato PNG de tamaño (640x640)
* file.txt. archivo con la descripción de los nombres de archivo de imagen y el peso del cuy.

### División del dataset en conjunto de entrenamiento y validación

```python
import os
from sklearn.model_selection  import train_test_split
import pandas as pd
import shutil
from PIL import Image
```

```python
# Leer el archivo de texto
df = pd.read_csv('C:\\Users\\Yeshica\\Pictures\\dataset\\cuy_bw.txt', header=None)
print('Hay {} imagenes con sus respectivos pesos'.format(len(df)))
```

Dividir el documento de texto en uno para el entrenamiento (80%) y otro para la validación (20%). Y los guardamos en la ruta indicada en formato .txt

```python
train, test = train_test_split(df, test_size=0.2, random_state=7)
train.to_csv(r'C:\\Users\\Yeshica\\Pictures\\dataset\\cuy_train.txt', header=None, index=None, sep=' ', mode='w')
test.to_csv(r'C:\\Users\\Yeshica\\Pictures\\dataset\\cuy_test.txt', header=None, index=None, sep=' ', mode='w')
```

```python
# Rutas donde se almacenan las imágenes y las máscaras
path_base = 'C:\\Users\\Yeshica\\Pictures\\dataset\\'
path_image_from = path_base + 'images'
path_mask_from = path_base + 'masks'
```

#### Training data

```python
# rutas para almacenar las imágenes y máscaras de entrenamiento
path_image_to = path_base + 'train_jpg'
path_mask_to = path_base + 'train_mask'
# Creamos las carpetas
os.mkdir(path_image_to)
os.mkdir(path_mask_to)
# Leemos el archivo con los datos de entrenamiento
file = open('C:\\Users\\Yeshica\\Pictures\\dataset\\cuy_train.txt')
lines = file.readlines()
for line in lines:
    filename = line.split(' ')[0]
    shutil.copy(os.path.join(path_image_from, filename), os.path.join(path_image_to, filename))
    maskname = filename.split('.')[0] + '_mask.png'
    shutil.copy(os.path.join(path_mask_from, maskname), os.path.join(path_mask_to, maskname))
```

Convertir de formato las imágenes (JPG a PNG), porque el aumentado se realiza en este formato.

```python
path_from = 'C:\\Users\\Yeshica\\Pictures\\dataset\\train_jpg'
path_to = 'C:\\Users\\Yeshica\\Pictures\\dataset\\train_png'
os.mkdir(path_to)
image_file_list = os.listdir(path_from)
for image_file in image_file_list:
    im = Image.open(os.path.join(path_from, image_file))
    filename =  os.path.splitext(os.path.basename(image_file))[0] + '.png'
    im.save(os.path.join(path_to, filename))
```

Luego, debemos reemplazar todas las extensiones (.jpg) del archivo .txt para continuar con el aumento de datos. Para el aumento de datos se necesita la ruta de entrada que contenga dos carpetas (images y masks con las imágenes y máscaras respectivamente), además de una carpeta de salida existente, el número de aumentaciones y el archivo donde se estan los nombres de archivos y pesos.
